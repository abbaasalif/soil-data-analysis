```python
import pandas as pd
```

```python
df = pd.read_csv('soilData.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
df['datetime'] = df['dates'].astype(str) + ' ' + df['hours'].astype(str) + ':00:00'
```

```python
df.drop(['dates', 'hours'], axis=1, inplace=True)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
df.head()
```

```python
import pytz
utc_tz = pytz.timezone('UTC')
local_tz = pytz.timezone('America/New_York')
df['local_datetime'] = df['datetime'].apply(lambda x: pd.Timestamp(x, tz=utc_tz).tz_convert(local_tz).tz_localize(None))
```

```python
xls = pd.ExcelFile('02-11-2023.xlsx')
df_radon={}
for key in ['48','9','1']:
    df_radon[key] = pd.read_excel(xls, key)
    df_radon[key] = df_radon[key]
```

```python
df_radon['1']
```

```python
df_radon_1 = df_radon['1'][['SyncDate','Temperature','Humidity','AirPressure','Radon']]
```

```python
df_radon_1['SyncDate'] = pd.to_datetime(df_radon_1['SyncDate'])
df_radon_1.sort_values(by='SyncDate', ascending=True, inplace=True)
df_radon_1['SyncDate'] = df_radon_1['SyncDate'].dt.floor('H')
df_radon_1 = df_radon_1.resample('H', on = 'SyncDate').mean()
df_radon_1 = df_radon_1.interpolate(method='linear', limit_direction='both')
df_radon_1.drop_duplicates(inplace=True)
```

```python
df_radon_1
```

```python
df['SyncDate'] = df['local_datetime']
```

```python
df
```

```python
df.drop(['datetime'], axis=1, inplace=True)
df.set_index('SyncDate', inplace=True)
```

```python
df
```

```python
for i in df_radon_1.columns:
    df[i] = df_radon_1[i]
```

```python
df
```

```python
#weather data
weather = pd.read_csv('weather.csv')
weather.drop('Date', axis=1, inplace=True)
weather['Simple Date'] = pd.to_datetime(weather['Simple Date'])
weather.sort_values(by='Simple Date', ascending=True, inplace=True)
weather = weather.resample('H', on = 'Simple Date').mean()
weather = weather.interpolate(method='linear', limit_direction='both')
weather.drop_duplicates(inplace=True)
```

```python
weather
```

```python
for i in weather.columns:
    df[i] = weather[i]
```

```python
df
```

```python
df.drop('local_datetime', axis=1, inplace=True)
```

```python
for column in df.columns:
    df[column] = df[column].astype('float32')
    if column == 'soil-moisture':
        continue
    elif column == 'soil-temperature':
        continue
    else:
        for i in range(1,241):
            df[column+'_lag_'+str(i)] = df[column].shift(i)
```

```python
# check correlations of features with price
df_corr = df.corr(method="pearson")
print(df_corr.shape)
print("correlation with soil-temperature:")
df_corrP = pd.DataFrame(df_corr["soil-temperature"].sort_values(ascending=False))
df_corrP
```

```python
import numpy as np
# highest absolute correlations with price
pd.options.display.float_format = '{:,.2f}'.format
df_corrH = df_corrP[(df_corrP["soil-temperature"] >= 0.80)]
df_corrH
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# correlation matrix, limited to highly correlated features
df3 = df[df_corrH.index]

idx = df3.corr().sort_values("soil-temperature", ascending=False).index
df3_sorted = df3.loc[:, idx]  # sort dataframe columns by their correlation with Appliances

plt.figure(figsize = (15,15))
sns.set(font_scale=0.75)
ax = sns.heatmap(df3_sorted.corr().round(3), 
            annot=False, 
            square=True, 
            linewidths=.75, cmap="coolwarm", 
            fmt = ".2f", 
            annot_kws = {"size": 11})
ax.xaxis.tick_bottom()
plt.title("correlation matrix")
plt.show()
```
